```python
import uproot
import awkward as ak
import numpy as np

# Correct branch names with / separators
plane_branches = [
    "trk/hits0/hits0.h.plane",
    "trk/hits1/hits1.h.plane", 
    "trk/hits2/hits2.h.plane"
]

# Load plane data
data = tree.arrays(plane_branches, library="ak")

print("Plane entry counts:")
print("="*60)

# Count for each hit collection
for i, branch in enumerate(plane_branches):
    print(f"\ntrk.hits{i}.h.plane:")
    planes_flat = ak.flatten(data[branch])
    unique_planes, counts = np.unique(planes_flat, return_counts=True)
    
    for p, count in zip(unique_planes, counts):
        print(f"  Plane {p}: {count} hits")

# Summary across all collections
print("\n" + "="*60)
print("Total hits per plane (across all collections):")
all_plane_counts = {}
for branch in plane_branches:
    planes_flat = ak.flatten(data[branch])
    unique_planes, counts = np.unique(planes_flat, return_counts=True)
    for p, count in zip(unique_planes, counts):
        all_plane_counts[p] = all_plane_counts.get(p, 0) + count

for plane_num in sorted(all_plane_counts.keys()):
    print(f"  Plane {plane_num}: {all_plane_counts[plane_num]} hits")
```
